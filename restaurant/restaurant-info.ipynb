```python
import pandas as pd
```

```python
data = pd.read_json('amenities-vancouver.json', lines = True)
```

```python
data
```

```python
food = data[(data.amenity == 'fast_food')|(data.amenity == 'restaurant')]
```

```python
food
```

```python
food = food.groupby('name').count()
```

```python
food
```

```python
food['chain'] = (food.tags >= 2)
```

```python
Chain = food[food.chain == True]
```

```python
NotChain = food[food.chain == False]
```

```python
NotChain
```

```python
Chain
```

```python
chain = pd.concat([Chain, NotChain])
```

```python
chain
```

```python
chain = chain.reset_index()
```

```python
chain
```

```python
chain = chain[["name", "chain"]]
```

```python
chain
```

```python
data = pd.merge(data, chain, on = "name")
```

```python
data
```

```python
food = data
```

```python
food
```

```python
food['NewChain'] = (food.tags >= 2)
```
